```python
import pandas as pd
from pandas_datareader import wb

# Indicators
indicators = {
    'NY.GDP.MKTP.CD': 'gdp',
    'GC.DOD.TOTL.GD.ZS': 'debt_to_gdp'
}

# Fetch data (all countries)
df = wb.download(
    indicator=list(indicators.keys()),
    country='all',
    start=2000,
    end=2023
)

# Clean up
df = df.rename(columns=indicators).reset_index()

print(df)
```

```python
new_df = df.iloc[:,[0,2]]
new_df  # 356 countries and 24 years (2000-2023)
```

```python
import matplotlib.pyplot as plt

# Feltételezzük: 0. oszlop = Országnév, 1. oszlop = Érték
# A groupby az ország neve alapján szétválogatja a 24 éves blokkokat
for country, group in new_df.groupby(new_df.columns[0]):
    plt.plot(range(2000, 2024), group.iloc[:, 1], label=country)

plt.title('Országok mutatói (24 év)')
plt.xlabel('Év')
plt.ylabel('Érték')

# Figyelem: 356 országnál a legend (jelmagyarázat) használhatatlan, ezért kikapcsoljuk
# plt.legend() 

```
